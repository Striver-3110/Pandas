# Series In pandas (1D data Structure)

```python
import pandas as pd 
#it contains 3 types of data types 
#1. Series 
#2. DataFrame 
#3. Panel
x = [3,4,5,6]
# x = [
#     [1,2,3,4],
#     [2,3,4,5],
#     [4,3,2,1],
#     [1,2,3,4]
# ]# no matter you pass 1d array or 2 d array ! pandas will convert it into one dimensional only
variable = pd.Series(x)
variable
variable = pd.Series(x,index=['a','b','c','d'],dtype="long")
#dtype = "int"is same as "long","float", "string", ,"byte"=int8,"short" = int 18 , "double" = "float64"but double will be consideres as float only
print(variable)
print(type(variable))

var = pd.Series(x,index=['a','b','c','d'],dtype="float",name="panda series")
print(var)
print(var.dtype)
print(var.name)
print(type(var))
```

```python
dic ={
    "name":['jay','c','c++'],
    "sir":['z.h.noorani','maitril shah','pragnesh patel']
}

pandaVar = pd.Series(dic)
print(pandaVar)
print(pandaVar["name"])
print(pandaVar["sir"][1])
manualIndexing = pd.Series(pandaVar["sir"],index= ['h','i','j'],dtype= "string",name="manualIndexing")
print(manualIndexing.dtype)
print(manualIndexing.index)
print(manualIndexing.name)
```

```python
var = 12
pandaSingleVar = pd.Series(var)
print(var)

pandaMultipleCopyOfSingleVar = pd.Series(var,index=[i for i in range(5)],dtype="int",name="pandaMultipleCopyUsingSingleVar")
print(pandaMultipleCopyOfSingleVar)
```

```python
#now lets perform addition bw two series

a1 = (1,2,3,4,5)
a2 = [9,8,7,6]

a1 = pd.Series(a1)
a2 = pd.Series(a2)
# a1 = pd.Series(a1)

print(a1+a2)

#where as numpy will throw an error in such cases where two list dont have same length
#lets see illustration

# import numpy as np

# a = [1,2,3]
# b = [1,2]
# a = np.array(a)
# b = np.array(b)
# print(a+b)
##################################################################

# this error will be generated by numpy array

# ----------------------------------------------------------------
# ValueError                     Traceback (most recent call last)
# Cell In[54], line 20
#      18 a = np.array(a)
#      19 b = np.array(b)
# ---> 20 print(a+b)

# ValueError: operands could not be broadcast together with shapes (3,) (2,) 


##################################################################
```

# DataFrame In pandas (1D data Structure)

```python
import pandas as pd

var = pd.DataFrame([1,2,3,4,5])
print(var)
print(type(var))
```

```python
lst = [
    [1,2,3,4],
    ['1','2','3','4'],
    [6,5,4] # in case of list, hetrogeneous length of list are allowed but in case of dictionaries it is not allowed
]


var = pd.DataFrame(lst)

print(var)
```

```python
d = {
    'a':[1,2,3,4],
    'b':[5,4,3,2],
#    'c':[1,2,4], will generate an error as length at each index of dic must be same
    'c':['a','s','f',1]
    
}

var = pd.DataFrame(d)
print(var)

# access perticular column

print(var['a'])

# accessing perticular value

print(var['c'][2])
```

```python
d = {
    'a':[1,2,3,4],
    'b':[5,4,3,2],
#    'c':[1,2,4], will generate an error as length at each index of dic must be same
    'c':['a','s','f',1]
}
var = pd.DataFrame(d,columns=['a',"c"],index=['a','s','e','r'])
# var = pd.DataFrame(d,columns=['a',"c"]) this will give auto index from 0 to n-1 where n = length of column
print(var)
```

```python
#remember that when multiple Series are combined, forms dataFrame
```

```python
a = [1,2,3,4]
b = ['a','m','t','r']
c = ["hii","my",1,5]

var = {
    "1" : pd.Series(a,index=['1',3,7,'jay']),
    '2' : pd.Series(b,index=['1',3,7,'jay']),
    '4' : pd.Series(c,index=['1',3,7,'jay'])
################################################
#     this will create dataframe with specified index
#     "1" : pd.Series(a,index=['-','=','+','@']),
#     '2' : pd.Series(b,index=['-','=','+','@']),
#     '4' : pd.Series(c,index=['-','=','+','@'])
#      }
#    1  2    4
# -  1  a  hii
# =  2  m   my
# +  3  t    1
# @  4  r    5
################################################
}
var = pd.DataFrame(var)
print(var)

# var1 = pd.DataFrame(var,columns=["1",'4'],index=[5,6,7,8])in correct way to create dataframe with index
# the correct one is given below see step 1 and step 2

# # step1
# var1 = pd.DataFrame(var,columns=["1",'4'])
# # step2
# var1.index = [1,2,3,4]

# print(var1)
```

# arithmetic operations in pandas

```python
dic = {
    1:[1,2,3,4],
    2:[5,4,3,2]
}
var = pd.DataFrame(dic)
```

```python
var[3] = var[1] + var[2]
print(var)
```

```python
var[3] = var[1] - var[2]
print(var)
```

```python
var[3] = var[1] // var[2]
print(var)
```

```python
var[3] = var[1] * var[2]
print(var)
```

```python
var[3] = var[1] / var[2]
print(var)
```

```python
var[3] = var[1] % var[2]
print(var)
```

```python
var['a>=2 (t or f)'] = var[1] >=2
print(var)
```

```python
var['a<4 (t or f)'] = var[1] <4
print(var)
```

```python
# (where as xls files are form of excel sheets in binary formate which holds information about all the 
# worksheets in a file, including both content and formatting)
```

```python
import pandas as pd
dic = {
    "a":[1,2,3,4,5,6],
    "b":[6,5,4,3,2,1],
    "d":[5,6,7,8,8,9]
}
d = pd.DataFrame(dic)

print(d)
```

```python
#how to create csv

d.to_csv("csvFirst.csv")# csv with index and header given by dic
d.to_csv("csvSec.csv",index=False)# csv without index but given header by dic
d.to_csv("csvThird.csv",index=False,header=["roll","maths","science"])# csv wihtout index and with user defined hader

# to check created files go to "C:\Pandas_tutorials"
```
